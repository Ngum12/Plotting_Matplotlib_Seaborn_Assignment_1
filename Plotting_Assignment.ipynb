# Histogram for Carbon Dioxide Emissions

```python
# YOUR CODE HERE
plt.hist(df['co2'], color='g', bins=40, edgecolor='black', alpha=0.5)
plt.title('Carbon Dioxide Distribution')
plt.xlabel('CO2 Emissions (g/mi)')
plt.ylabel("Frequency")
plt.show()
```

# Experiment with Different Bin Sizes

```python
# Example with different bin sizes
plt.hist(df['co2'], color='g', bins=20, edgecolor='black', alpha=0.5)
plt.title('Carbon Dioxide Distribution (20 Bins)')
plt.xlabel('CO2 Emissions (g/mi)')
plt.ylabel("Frequency")
plt.show()

plt.hist(df['co2'], color='g', bins=60, edgecolor='black', alpha=0.5)
plt.title('Carbon Dioxide Distribution (60 Bins)')
plt.xlabel('CO2 Emissions (g/mi)')
plt.ylabel("Frequency")
plt.show()
```

# Histogram of CO2 Emissions
**Description**

Distribution Shape: The histogram typically shows a bell-shaped curve, indicating that CO2 emissions are approximately normally distributed.
Peak Range: The peak around 300-400 g/mi suggests that a majority of vehicles emit CO2 within this range, indicating a common efficiency level.
Implications: This normal distribution implies that most vehicles are designed to meet similar emission standards, but there are fewer vehicles at the extremes (very low or very high emissions). This can inform regulatory bodies about the effectiveness of emission standards.

# Histogram of Combined Fuel Efficiency

```python
# YOUR CODE HERE
# Histogram for Combined Fuel Efficiency
plt.figure(figsize=(10, 6))
sns.histplot(df['comb'], bins=30, kde=False)
plt.title('Distribution of Combined Fuel Efficiency')
plt.xlabel('Combined Fuel Efficiency (mpg)')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

# Histogram of Combined Fuel Efficiency

**Description**

Distribution Shape: This histogram often appears right-skewed, indicating that while many vehicles have lower fuel efficiencies, there are some that achieve significantly higher efficiencies.
Peak Range: The peak around 20-25 mpg indicates that most vehicles fall within this efficiency range, but the long tail towards higher efficiencies suggests the presence of more efficient models.
Implications: The skewness indicates potential market opportunities for manufacturers to develop more fuel-efficient vehicles, as there is a demand for better fuel economy.

# Histogram for Highway Fuel Efficiency

```python
# YOUR CODE HERE
plt.figure(figsize=(10, 6))
sns.histplot(df['highway'], bins=30, kde=False)
plt.title('Distribution of Highway Fuel Efficiency')
plt.xlabel('Highway Fuel Efficiency (mpg)')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

# Histogram of Highway Fuel Efficiency

**Description**

Distribution Shape: Similar to the combined fuel efficiency histogram, the highway fuel efficiency distribution may also show a right skew.
Peak Range: The peak typically appears in the mid-20s mpg, indicating that this is a common efficiency level for highway driving.
Implications: The data suggests that while many vehicles perform adequately on highways, there is still a gap for improvement in fuel efficiency, particularly for larger vehicles or those with higher emissions.

# Create Boxplots
Generate boxplots to visualize the distribution and identify outliers for CO2 emissions and combined fuel efficiency.

```python
# Boxplot for CO2 emissions
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['co2'])
plt.title('Boxplot of CO2 Emissions')
plt.xlabel('CO2 Emissions (g/mi)')
plt.grid()
plt.show()

# Boxplot for Combined Fuel Efficiency
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['comb'])
plt.title('Boxplot of Combined Fuel Efficiency')
plt.xlabel('Combined Fuel Efficiency (mpg)')
plt.grid()
plt.show()
```

# **Columns to Drop for GHG Score Prediction**
When focusing on predicting the GHG score, certain columns may be deemed less relevant and can be dropped to streamline the analysis.

**Columns to Drop:**

# Make and Model:
These categorical variables do not provide numerical value in predicting GHG scores and can introduce noise into the model.
# Year:
While it may seem relevant, the year alone does not directly correlate with GHG emissions unless contextualized with other factors like technological advancements in fuel efficiency.
# VClass:
Vehicle class may not significantly impact GHG scores directly, as the GHG score is more dependent on the vehicle's fuel consumption and emissions rather than its classification.
# Drive, Trans, FuelType:
While these can have some influence, they may not be as critical as numerical factors like cylinders, displacement, and fuel economy metrics.
# **Justification:**
Dropping these columns reduces dimensionality, which can help improve model performance and interpretability.
It focuses the analysis on quantitative attributes that have a more direct impact on GHG emissions, such as fuel efficiency metrics and engine specifications.
